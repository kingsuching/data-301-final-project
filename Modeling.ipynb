```python
from re import A
from sklearn.preprocessing import *
from sklearn.model_selection import *
from sklearn.compose import *
from sklearn.pipeline import *
from sklearn.neighbors import *
from sklearn.compose import *
from sklearn.cluster import *
means = df_departures.groupby("origin.code")["departure_delay"].mean()
arr = []
for airport in df["airport_code"]:
  arr.append(means[airport])
df["delay_secs"] = arr
```

```python
def trainer(X_train, y_train, k = 5, pred = False):
  transformer = make_column_transformer(
    (OneHotEncoder(handle_unknown="ignore"), make_column_selector(dtype_exclude = np.number)),
    (StandardScaler(), make_column_selector(dtype_include = np.number))
  )
  pipeline = make_pipeline(transformer, KNeighborsRegressor(k))
  pipeline.fit(X_train, y_train)
  if pred:
    return pipeline.predict(X_train)
  return pipeline
```

```python
df["altitude"] = df["altitude"].astype(float)
X_train_df = df.drop(["airport_code", "pressure_units", "visibility_units", "wind_units", "symbol", "raw_data", "time", "delay_secs"], axis=1)
y_train_df = df["delay_secs"].fillna(0) # departure delay
y_train_df.index = range(len(y_train_df))
for col in X_train_df.columns:
  if X_train_df.dtypes[col] == object:
    X_train_df[col].fillna("", inplace=True)
  else:
    X_train_df[col].fillna(0, inplace=True)

pipeline = trainer(X_train_df, y_train_df)
predictions = pipeline.predict(X_train_df)
abs(y_train_df - predictions).mean()
np.sqrt(-cross_val_score(pipeline, X_train_df, y_train_df, scoring="neg_mean_squared_error", cv=10)).mean()
abs(y_train_df - pd.Series(predictions)).mean()/60
```

```python
def best_k(X_train, y_train, k=30):
  model = trainer(X_train, y_train)
  grid = GridSearchCV(model, param_grid={
                               "kneighborsregressor__n_neighbors": range(2, k+1)
                           }, scoring="neg_mean_squared_error", cv=10)
  grid.fit(X_train, y_train)
  return grid.best_estimator_

#best_k(X_train_dep, y_train_dep)
best_k(X_train_arr, y_train_arr)
```

```python
def modelor(X_train, y_train, k=5):
  transformer = make_column_transformer(
      (OneHotEncoder(handle_unknown="ignore"), make_column_selector(dtype_exclude = np.number)),
      (StandardScaler(), make_column_selector(dtype_include = np.number))
  )
  pipeline = make_pipeline(transformer, KNeighborsClassifier(n_neighbors=k))
  pipeline.fit(X_train, y_train)
  return pipeline
```
